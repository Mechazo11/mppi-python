# Script to ensure DiffDrive model works, animation viewing and saving works

```python
# Imports
import math
import numpy as np
from scripts.diffdrive import DiffDriveVehicle
from scripts.animation import Animator
from scripts.utils import curr_time

sim_step = 50
delta_t = 0.1 # Time step
ref_path = np.genfromtxt('./data/ovalpath.csv', delimiter=',', skip_header=1)
vehicle = DiffDriveVehicle(delta_t=delta_t,ref_path=ref_path[:, 0:2])

animator = Animator(vehicle_model=vehicle)
for i in range(sim_step):
    tt = curr_time()
    # u is the control input to the vehicle, [ steer[rad], accel[m/s^2] ]
    u_sample = [0.5, 0.0] # Go front
    #u_sample = [-1.0, 0.0] # Go back
    #u_sample = [0.0, 0.5] # CCW rotation
    # vehicle.update(u=[2.0 * np.sin(i/5.0), -2.0 * np.sin(i/10.0)], delta_t=delta_t) 
    vehicle.update(u=u_sample, x_t_prev=vehicle.get_state(), delta_t=delta_t) 
    animator.draw_frame(vehicle.out_arr)
    # print(f"Timestep --> {i} took: {curr_time() - tt} ms to finish")

animator.show_animation(interval_ms=delta_t*1000) # show animation
# animator.save_animation("test.mp4", interval=int(delta_t * 1000), movie_writer="ffmpeg")
```
